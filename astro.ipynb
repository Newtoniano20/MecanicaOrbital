```python
from sympy import plot_implicit, Eq
from sympy.abc import x, y
from sympy.plotting.plot import MatplotlibBackend, Plot
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots()
```

```python
a = 0.0020819929702443124
e = 1.3210621781218228
b = -a**2 + (e*a)**2
b = np.sqrt(b)
```

```python
f = lambda x: -np.sqrt((- 1 + (x/a)**2)*b**2)
```

```python
x = np.arange(-0.005, 0.005, 0.0000001)
y = [f(_x) for _x in x]
```

```python
plt.plot(y, x)
plt.scatter(-1, 0)
plt.xlim(-0.005, 0.005)
plt.ylim(-0.005, 0.005)
```

```python
plot_implicit(Eq((), (x, -0.1, 0.00), (y, -0.032, 0.040))
```
